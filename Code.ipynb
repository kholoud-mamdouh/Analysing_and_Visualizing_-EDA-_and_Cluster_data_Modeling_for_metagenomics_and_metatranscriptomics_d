#*Encoding data*


```python
data_modeling=df.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Initialize the OneHotEncoder
encoder = OneHotEncoder()

# Create a copy of the dataframe
data_modeling_encoded = data_modeling.copy()

# Specify the columns to be one-hot encoded
columns_to_encode = [" MG unprocessed ", " MG retained pairs ", " MG retained singles ", " MT contigs ", " MT retained pairs ", " MT retained singles "]

# Fit and transform the specified columns
for column in columns_to_encode:
    data_temp = data_modeling[[column]]
    encoder.fit(data_temp)
    temp = encoder.transform(data_temp).toarray()
    temp = pd.DataFrame(temp, columns=[(column + "_" + str(i)) for i in encoder.categories_[0]])
    data_modeling_encoded = data_modeling_encoded.drop(column, axis=1)
    data_modeling_encoded = data_modeling_encoded.join(temp)
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import numpy as np
pca = PCA(2)
df = pca.fit_transform(mod_data)
df.shape
```

```python
# Print the explained variance ratio of the first two components
print("Explained variance ratio:", pca.explained_variance_ratio_)

# Create a new dataframe from the PCA transformed data
data_modeling_pca = pd.DataFrame(df, columns=['PC1', 'PC2'])
```

```python
from sklearn.cluster import KMeans

# Initialize the KMeans model with 4 clusters
kmeans = KMeans(n_clusters=4)

# Fit the model to the PCA transformed data
kmeans.fit(df)

# Get the cluster assignments for each sample
clusters = kmeans.predict(df)

# Add the cluster assignment as a new column to the original dataframe
data_modeling_pca['cluster'] = clusters

# Plot the data points for each cluster
import matplotlib.pyplot as plt
for i in range(4):
    cluster = data_modeling_pca[data_modeling_pca['cluster'] == i]
    plt.scatter(cluster['PC1'], cluster['PC2'])

# Add axis labels and show the plot
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```

```python
import seaborn as sns

# Plot the violin plots
sns.violinplot(data=data_modeling_pca, x='cluster', y='PC1', inner='stick')
sns.violinplot(data=data_modeling_pca, x='cluster', y='PC2', inner='stick')

# Add axis labels and show the plot
plt.xlabel('Cluster')
plt.ylabel('PCA Transformed Data')
plt.show()
```
